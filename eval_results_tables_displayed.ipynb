Displaying results tables from evaluation of the models

```python
import pandas as pd
from IPython.display import display

def print_line(char, times):
    print(char * times)
    
# Load JSON files directly into DataFrames
when_df = pd.read_json('results/when.json')
where_df = pd.read_json('results/where.json')
who_df = pd.read_json('results/who.json')

width = 800
# Adjust Pandas settings for better display
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', width)       # Set display width
pd.set_option('display.max_colwidth', None)  # Show full column content

# Display the DataFrames
# When Data
print_line('-', width)
print("When Data:")
print_line('-', width)
display(when_df)

# Where Data
print_line('-', width)
print("Where Data:")
print_line('-', width)
display(where_df)

# Who Data
print_line('-', width)
print("Who Data:")
print_line('-', width)
display(who_df)
```
